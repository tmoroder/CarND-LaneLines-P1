## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

# Solution
In the following we stitch together the pipeline with the above defined helper functions. The pipeline logic is the same as explained in the course material. It consists of the following steps:
1. Edge detection
  - Convert image to grayscale
  - Gaussian blur
  - Canny edge detection
2. Selection of region of interest 
3. Line detection using Hough transformation
4. Draw detected lines on image

## Comments
As emphasized in the course, a large portion consists in parameter tuning. To this end, I employ Juyter widgets in this notebook to interactively carry out this optimization. However, this is not visible in the static version of the notebook; a screenshot is included in the writeup. The results are included in a configuration object.

````python
# Imports

from collections import namedtuple
import ipywidgets as widgets
````

````python
# Utility functions

def show_images(d_images, cols=2):
    """Displays all images in the dictionary."""
    rows = -(-len(d_images) // cols)
    fig = plt.figure(figsize=(16, 4 * rows))
    for i, k in enumerate(d_images, start=1):
        plt.subplot(rows, cols, i)
        img = d_images[k]
        cm = "gray" if img.ndim == 2 else None
        plt.imshow(img, cmap=cm)
        plt.axis('off')
        plt.title(k)
    plt.tight_layout()
    return fig
````

````python
# Pipeline function with corresponding Config object

Config = namedtuple("Config", (
    " kernel_size low_threshold high_threshold " 
    " vert_height, vert_delta_width "
    " rho theta_degree threshold min_line_len max_line_gap "
    ),
)


def get_vertices(vert_height, vert_delta_width, height, width):
    """Gets vertices from relative height and width."""
    return np.array([[
        [0, height], 
        [(0.5 - vert_delta_width) * width, vert_height * height], 
        [(0.5 + vert_delta_width) * width, vert_height * height],
        [width, height]
    ]], dtype=np.int)


def pipeline(img, cfg: Config, return_collection=False):
    """Pipeline function detecting lanes in the images. Parameters of the 
    individual pipeline functions are takes from passed 'Config' object. 
    Optionally returns collection of internal images used in interactive
    parameter tuning and pipeline documenation.
    """
    # edge detection: gray - blur - canny
    img_gray = grayscale(img)
    img_blur = gaussian_blur(img_gray, kernel_size=cfg.kernel_size)
    img_edges = canny(
        img_blur, low_threshold=cfg.low_threshold, high_threshold=cfg.high_threshold
    )
    # region of interest
    vertices = get_vertices(cfg.vert_height, cfg.vert_delta_width, *img.shape[:2])
    img_roi_edges = region_of_interest(img_edges, vertices)
    # line detection
    theta = np.deg2rad(cfg.theta_degree)
    img_lines = hough_lines(
        img_roi_edges, cfg.rho, theta, cfg.threshold, cfg.min_line_len, cfg.max_line_gap
    )
    # add lines to image
    img_result = weighted_img(img_lines, img)
    if return_collection:
        img_roi_result = region_of_interest(img_result, vertices)
    # return result
    result = img_result
    if return_collection:
        result = {
            "original": img,
            "edges": img_edges,
            "roi_edges": img_roi_edges,
            "lines": img_lines,
            "result": img_result,
            "roi_result": img_roi_result
        }
    return result
````

````python
# Parameter tuning
# Initial values are now set to my chosen solution

CFG = None
IMG_FOLDER = "./test_images/"


@widgets.interact(
    image_name=widgets.Dropdown(options=list(os.listdir(IMG_FOLDER))),
    kernel_size=widgets.IntSlider(value=5, min=1, max=11, step=2, continuous_update=False),
    low_threshold=widgets.IntSlider(value=68, min=0, max=255, continuous_update=False),
    high_threshold=widgets.IntSlider(value=150, min=0, max=255, continuous_update=False),
    vert_height=widgets.FloatSlider(
        value=0.61, min=0.0, max=1.0, step=0.01, continuous_update=False
    ),
    vert_delta_width=widgets.FloatSlider(
        value=0.05, min=0.0, max=0.5, step=0.01, continuous_update=False
    ),
    rho=widgets.IntSlider(value=1, min=1, max=15, step=1, continuous_update=False),
    theta_degree=widgets.IntSlider(value=1, min=1, max=45, continuous_update=False),
    threshold=widgets.IntSlider(value=20, min=1, max=50, continuous_update=False),
    min_line_len=widgets.IntSlider(value=42, min=1, max=50, continuous_update=False),
    max_line_gap=widgets.IntSlider(value=35, min=1, max=50, continuous_update=False),
)
def _parameter_tuning(**kwargs):
    # set config
    global CFG
    CFG = Config(**{k: v for k, v in kwargs.items() if k in Config._fields})
    # load image
    img_path = os.path.join(IMG_FOLDER, kwargs["image_name"])
    img_orig = mpimg.imread(img_path)
    img = np.copy(img_orig)
    # call pipeline with current config and return collection of images
    d_images = pipeline(img, CFG, return_collection=True)
    # show
    fig = show_images(d_images, 3)
    plt.show()
````

# Solution
The following method is employed to extend the line segments over the full visible lane:

1. Divide line segments into left and right lane segments, for which the slope of the line segment is used.
2. Average slope and intercept of all line segments per side
3. Compute end points at region of interest heights for each side
4. Draw those lines to the image

## Comments
- A weighted average using line length is employed so that longer lines have a stronger effect.
- Only line segments with slope corresponding to an (absolute) angle in [20, 70] are considered for averaging. A line segment outside this interval is either too flat or too steep to correspond to the lane line. Note, this also avoids problems with very extreme slope values.
- The following modifications were tried but with no, or only a negligible effect, regarding robustness:
  * Median (plain or weighted) instead of averging.
  * Polar coordinates rather than (slope, intercept) pairs.
- As an implementation detail, the variable ``draw_lines`` is assigned to a newly function, so that above ``pipeline`` function can be re-used.

## Development

````python
# Variable 'draw_lines_orig' is assigned to old draw_lines function
# Below the 'draw_lines' variable will be assigned to new function

if "draw_lines_orig" not in locals():
    print("draw_lines saved")
    draw_lines_orig = draw_lines
````

````python
# Helper functions

def compute_line_parameter(x1, y1, x2, y2):
    """Compute (slope, intercept and length) of line."""
    s = np.divide((y2 - y1), (x2 - x1))
    b = y1 - s * x1
    w = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return np.array([s, b, w], np.float)


def slope2degree(x):
    """Slope to degree"""
    return np.rad2deg(np.arctan(x))
````

````python
# Process lines and new draw_lines function

BORDER_DEGREE = [20., 70.]
ROI_HEIGHT = CFG.vert_height


def process_lines(lines, img_height):
    # Compute line parameter (slope, intercept and length), remove too flat/steep
    # lines and divide into left and right lane depending on slope
    left, right = [], []
    for line in lines:
        line_param = compute_line_parameter(*line[0])
        slope = line_param[0]
        abs_theta = np.abs(slope2degree(slope))
        if BORDER_DEGREE[0] <= abs_theta <= BORDER_DEGREE[1]:
            if slope < 0:
                left.append(line_param)
            else:
                right.append(line_param)
    # Average slope and intercept, and compute end points at ROI heights for each side
    lines = []
    for lane in [left, right]:
        if not lane:
            continue
        # Average line parameters
        data = np.array(lane)
        slope, intercept = np.average(data[:, :2], weights=data[:, 2], axis=0)
        # Compute end points
        line = []
        for y in [int(ROI_HEIGHT * img_height), img_height]:
            x = np.int((y - intercept) / slope)
            line += [x, y]
        lines.append(np.array(line)[np.newaxis])
    return lines


def draw_lines_lanes(img, lines, color=[255, 0, 0], thickness=2):
    lines = process_lines(lines, img_height=img.shape[0])
    draw_lines_orig(img, lines, color, 6 * thickness)
````

````python
# Set new draw_lines function

draw_lines = draw_lines_lanes
````
